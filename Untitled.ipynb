```python
df = quandl.get("EURONEXT/ABN")
df_filled = df.asfreq("D", method = "ffill") #this is to fill the weekend will the previouse weekday data
df_pct_change = df_filled.pct_change()
returns = df_pct_change["Last"]
df_return_std = returns.rolling(window = 30, min_periods = 30).std()
df_return_std.plot()
```
